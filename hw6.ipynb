```python
#1

class BTnode:

  def __init__(self, data, left = None, right = None):
    self.data = data
    self.right = right
    self.left = left

n = BTnode(1,BTnode(2),BTnode(3,None,BTnode(4)))


class BTree:

  def __init__(self, root = None):
    self.root = root

  # LNR order
  def inorder(self, root):
    if root == None :
      return
    else:
      self.inorder(root.left)
      print(root.data)
      self.inorder(root.right)


  #LRN
  def postorder(self,root):
    if root == None:
      return
    else:
      self.postorder(root.left)
      self.postorder(root.right)
      print(root.data)


  #NLR
  def priorder(self,root):
    if root == None:
      return
    else:
      print(root.data)
      self.priorder(root.left)
      self.priorder(root.right)



tree = BTnode(1,BTnode(2,BTnode(4),BTnode(5)),BTnode(3))

tree1 = BTree()
tree1.inorder(tree)
tree1.postorder(tree)
tree1.priorder(tree)
```

```python
#3

def heapify(A , k):
  left = 2*k+1
  right = 2*k+2
  if left < len(A) and A[left] < A[k]:
    s = 1
  else:
    s = k
  if right < len(A) and A[right] < A[s]:
    s = right
  if s!=k:
    A[s],A[k] = A[k],A[s]
    heapify(A , s)



#min heap

def build_min_heap(A):
  n = ((len(A)//2)-1)
  for k in range(n,-1,-1):
    heapify(A,k)


A=[5,6,3,1,4,2,7]
build_min_heap(A)
print(A)

```
